```python
%%sql
SELECT
Participant_Code,
Outage_Reason, Year,
COUNT(Outage_Reason) AS Total_Number_Outage_Events,
ROUND(AVG((ABS(JULIANDAY(End_Time)-JULIANDAY(Start_Time)))), 2) AS Average_Outage_Duration_In_Days
FROM AEMR_Outage_Table
WHERE Status='Approved'
GROUP BY Participant_Code, Outage_Reason, Year
ORDER BY Total_Number_Outage_Events DESC, Outage_Reason, Year
```

```python
%%sql
WITH Cte AS(
SELECT 
    Participant_Code, Outage_Reason, Year, 
    COUNT(Outage_Reason)AS Total_Number_Outage_Events, 
    ROUND(AVG(ABS(JULIANDAY(End_Time)-(JULIANDAY(Start_Time))),2) AS Average_Outage_Duration_In_Days
          FROM AEMR_Outage_Table
          WHERE Status='Approved'
GROUP BY Participant_Code, Outage_Reason, Year
ORDER BY Average_Outage_Duration_In_Days DESC
         )

SELECT 
    Participant_Code, Outage_Reason, Year, Total_Number_Outage_Events, Average_Outage_Duration_In_Days,
    CASE
    WHEN Average_Outage_Duration_In_Days<0.5 THEN 'Low Risk'
    WHEN Average_Outage_Duration_In_Days<1.0 THEN 'Medium Risk'
    ElSE 'High Risk'
    END AS Risk_Classification
    FROM Cte
GROUP BY Participant_Code, Outage_ Reason, Year, Risk_Classification
ORDER BY Total_Number_Outage_Events DESC, Average_Outage_Duration_In_Days DESC
```

```python
%%sql 
WITH Ctes_Clas AS(
SELECT 
Participant_Code, Outage_Reason, Year, COUNT(Outage_Reason) AS Total_Number_Outage_Events,
    ROUND(AVG(ABS(JULIANDAY(End_Time)-JULIANDAY(Start_Time))),2) AS Average_Outage_Duration_In_Days
    FROM AEMR_Outage_Table
    WHERE Status='Approved'
    GROUP BY Participant_Code, Outage_Reason, Year
    ORDER BY Average_Outage_Duration_In_Days DESC)
    
SELECT
Participant_Code, Outage_Reason, Year, Total_Number_Outage_Events, Average_Outage_Duration_In_Days,
CASE
WHEN Outage_Reason<>'Forced'THEN 'N/A'
WHEN (Outage_Reason='Forced' AND (Average_Outage_Duration_In_Days<0.5 OR Total_Number_Outage_Events<10) ) THEN 'Low Risk'
WHEN (Outage_Reason='Forces' AND (Average_Outage_Duration_In_Days<1 OR Total_Number_Outage_Events<20))THEN 'Medium Risk'
ELSE 'High Risk'
END AS Risk_Classification

FROM Ctes_Clas
GROUP BY Participant_Code, Outage_Reason, Year, Risk_Classification
ORDER BY Participant_Code, Outage_Reason DESC, Average_Outage_Duration_In_Days DESC
    
```

```python
%%sql
WITH Table_One AS(
SELECT Year, Outage_Reason, COUNT(Outage_Reason)AS Total_Number_Outages
    FROM AEMR_Outage_Table
    WHERE Status='Approved'
    GROUP BY Year),
    Table_Two AS (
    SELECT Year, Outage_Reason, COUNT (Outage_Reason) AS Total_Number_Forced_Outage_Events
        FROM AEMR_Outage_Table
        WHERE Outage_Reason='Forced'AND Status='Approved'
        GROUP BY Year)
        

SELECT 
Table_One.Year, Table_One.Outage_Reason, Total_Number_Outages, Total_Number_Forced_Outage_Events, 
ROUND((Total_Number_Forced_Outage_Events*100.00)/Total_Number_Outages,2) AS Pct_Outage_Forced
FROM Table_One
JOIN Table_Two ON Table_One.Year=Table_Two.Year
```

```python
%%sql
WITH Table_One AS(
SELECT
Participant_Code, Facility_Code, Description_Of_Outage, ROUND(SUM(Energy_Lost_MW),2) AS Total_Energy_Lost, Year
    FROM AEMR_Outage_Table
    WHERE Status='Approved' AND Participant_Code IN ('GW','MELK','AURICON')
    GROUP BY Participant_Code, Facility_Code
    ORDER BY Total_Energy_Lost DESC),
Table_Two AS(
SELECT 
    Participant_Code, Description_Of_Outage, ROUND(SUM(Energy_Lost_MW),2) AS Energy_Lost
FROM AEMR_Outage_Table
WHERE Status='Approved' AND Participant_Code IN ('GW','MELK','AURICON')
GROUP BY Description_Of_Outage
ORDER BY Energy_Lost DESC)
SELECT
Table_One.Participant_Code, Table_One.Facility_Code, Table_two.Description_Of_Outage, Total_Energy_Lost,
ROUND(Energy_Lost*100.00/Total_Energy_Lost,2)AS Pct_Energy_Loss, Rank() OVER(PARTITION BY Table_Two.Description_Of_Outage
                                                                            ORDER BY Table_Two.Participant_Code)AS rank

FROM Table_One
JOIN Table_Two ON Table_Two.Description_Of_Outage=Table_One.Description_Of_Outage
ORDER BY Total_Energy_Lost DESC
```
